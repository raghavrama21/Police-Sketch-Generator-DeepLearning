```python
import os
import shutil

def extract_files(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    # Iterate over all files in the source folder
    for filename in os.listdir(source_folder):
        source_path = os.path.join(source_folder, filename)
        destination_path = os.path.join(destination_folder, filename)

        # Check if the current item is a file
        if os.path.isfile(source_path):
            # Move the file to the destination folder
            shutil.move(source_path, destination_path)
            print(f"Moved file: {source_path} -> {destination_path}")

# Example usage:
source_folder = r'C:\Users\ragha\Desktop\cfd\CFD Version 3.0\Images\CFD-MR'
destination_folder = r'C:\Users\ragha\Desktop\cfd\CFD Version 3.0\Images\CFD-Overall'

extract_files(source_folder, destination_folder)
```

```python
import os

def get_folder_list(parent_folder):
    folder_list = []
    for item in os.listdir(parent_folder):
        item_path = os.path.join(parent_folder, item)
        if os.path.isdir(item_path):
            folder_list.append(item_path)
    return folder_list

# Example usage:
parent_folder =  r'C:\Users\ragha\Desktop\cfd\CFD Version 3.0\Images\CFD'
folders = get_folder_list(parent_folder)

print(folders)
```

```python
import numpy as np
import pandas as pd

def bucketize_column(data, column_name, num_buckets):
    # Calculate the bucket boundaries
    min_value = data[column_name].min()
    max_value = data[column_name].max()
    bucket_size = (max_value - min_value) / num_buckets

    return pd.cut(data[column_name], bins=num_buckets, labels=False)


data2 = pd.read_csv(r"C:\Users\ragha\Desktop\UC Davis MSBA 2022 - 2023\BAX - Spring Quarter 2023\Top Big Data Scientist\data2.csv")
list(data2.columns[3:])
```
